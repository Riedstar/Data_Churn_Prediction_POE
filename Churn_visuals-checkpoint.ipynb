```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 

# For inline plots

# Load sheets
overall = pd.read_excel('churn_results.xlsx', sheet_name='Overall Stats')
metrics = pd.read_excel('churn_results.xlsx', sheet_name='Week 05 Metrics')
model_perf = pd.read_excel('churn_results.xlsx', sheet_name='Model Performance')
segments = pd.read_excel('churn_results.xlsx', sheet_name='Segments')

print("Loaded! Churn rate:", overall['churn_rate'].iloc[0])
```

```python
## In this chart we will demonstrate how churn cutmers are divided vs retention

##Ocerall churn severity shows that 23k weekly droppers of data as revenue risk(R700k+ loss @R30-R60 ARPU)


churn_pct = overall['churn_rate'].iloc[0] * 100
retain_pct = 100 - churn_pct
plt.figure(figsize=(6,6))
plt.pie([churn_pct, retain_pct], labels=[f'Churn ({churn_pct:.1f}%)', f'Retention ({retain_pct:.1f}%)'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Weekly Data Churn Rate')
plt.savefig('churn_pie.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
## In this report my obervations are : Clustered bars comparing retainers vs. droppers on week 05,
## metrics (active days, sessions, volume in GB), with hue legend.


metrics_melt = pd.melt(metrics, id_vars='Target', var_name='Metric', value_name='Value')
metrics_melt['Group'] = metrics_melt['Target'].map({0:'Retainer',1:'Dropper'})
plt.figure(figsize=(10,6))
sns.barplot(data=metrics_melt[metrics_melt['Metric'] != 'count'], x='Group', y='Value', hue='Metric')
plt.title('Week 05 Engagement Gaps')
plt.ylabel('Avg Value')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.savefig('engagement_gaps.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
## In this report my obeservations are, Horizontal bar plot of F1 scores for 3 models (XGBoost highest at 0.750), with value annotations. 
## (Optional radar adds multi-metric view.)



plt.figure(figsize=(8,5))
sns.barplot(data=model_perf, x='F1 (Dropper)', y=model_perf.index, palette='viridis')
plt.title('Model F1 Scores: XGBoost Leads')
plt.xlabel('F1 (Dropper)')
for i, v in enumerate(model_perf['F1 (Dropper)']):
    plt.text(v + 0.01, i, f'{v:.3f}')
plt.savefig('model_f1_bar.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
## In this report we can observe that the Donut pie (white center) with user % by risk tier (15% high-risk red, 13% medium orange, 72% low green).
## This Represents: XGBoost-predicted segmentsfocuses targeting (e.g., 9K high-risk for campaigns)





plt.figure(figsize=(7,7))
plt.pie(segments['Users'], labels=segments['Segment'], autopct='%1.1f%%', colors=['#ff9999','#ffcc99','#99ff99'])
plt.title('Churn Risk Segments')
centre = plt.Circle((0,0),0.7,fc='white')
plt.gca().add_artist(centre)
plt.savefig('segments_donut.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
## Here we demonstrating with Side-by-side bars: Churn rate drop (38.4% red to 32% green) + weekly rev uplift (R140K blue bar).



before = 38.4
after = 32.0
rev_uplift = 140000
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,5))
ax1.bar(['Before','After'], [before,after], color=['red','green'])
ax1.set_ylabel('Churn Rate (%)')
ax1.set_title('Churn Reduction')
ax2.bar(['Weekly Uplift'], [rev_uplift/1000], color='blue')
ax2.set_ylabel('R\'000')
ax2.set_title('Revenue Boost')
plt.tight_layout()
plt.savefig('impact_bars.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cell 8: Line  Weekly Data Revenue Trends by Churn Status
# Aggregate weekly rev (avg TOT_DATA_REV_XX by Target)
# week_cols = ['TOT_DATA_REV_01', 'TOT_DATA_REV_02', 'TOT_DATA_REV_03', 'TOT_DATA_REV_04', 'TOT_DATA_REV_05']
# full_df['week'] = range(1, 6)
# rev_trends = full_df.groupby(['Target', 'week'])[week_cols].mean().reset_index()
# rev_trends = rev_trends.melt(id_vars=['Target', 'week'], var_name='col', value_name='rev')
# rev_trends['week_num'] = rev_trends['col'].str[-2:].astype(int)

# plt.figure(figsize=(10, 6))
# sns.lineplot(data=rev_trends, x='week_num', y='rev', hue='Target', marker='o', linewidth=2)
# plt.title('Weekly Data Revenue Trends (Avg by Churn Status)')
# plt.xlabel('Week (1=Oldest, 5=Recent)')
# plt.ylabel('Avg TOT_DATA_REV (R)')
# plt.legend(title='Target (1=Dropper)')
# plt.grid(True, alpha=0.3)
# plt.savefig('revenue_trends_line.png', dpi=300, bbox_inches='tight')
# plt.show()



# Cell 8: Line  Weekly Data Revenue Trends by Churn Status
# Aggregate weekly rev (avg TOT_DATA_REV_XX by Target)
week_cols = ['TOT_DATA_REV_01', 'TOT_DATA_REV_02', 'TOT_DATA_REV_03', 'TOT_DATA_REV_04', 'TOT_DATA_REV_05']

# Calculate mean revenue by Target group for each week column
rev_trends = full_df.groupby('Target')[week_cols].mean().reset_index()

# Reshape data from wide to long format for plotting
rev_trends = rev_trends.melt(id_vars=['Target'], value_vars=week_cols, 
                            var_name='week_col', value_name='rev')

# Extract week number from column names
rev_trends['week_num'] = rev_trends['week_col'].str[-2:].astype(int)

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=rev_trends, x='week_num', y='rev', hue='Target', marker='o', linewidth=2)
plt.title('Weekly Data Revenue Trends (Avg by Churn Status)')
plt.xlabel('Week (1=Oldest, 5=Recent)')
plt.ylabel('Avg TOT_DATA_REV (R)')
plt.legend(title='Target (1=Dropper)')
plt.grid(True, alpha=0.3)
plt.savefig('revenue_trends_line.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cell 9: Box  Data Bundles Bought by Risk Segment (Week 05)
full_df['risk_segment'] = pd.cut(full_df['proba_drop'], bins=[0, 0.4, 0.7, 1], labels=['Low', 'Medium', 'High'])

plt.figure(figsize=(8, 6))
sns.boxplot(data=full_df, x='risk_segment', y='DATA_DNDL_CNT_05', palette='Set2')
plt.title('Data Bundle Purchases by Risk Segment (Week 05)')
plt.xlabel('Risk Segment')
plt.ylabel('Count of Data Bundles Bought')
plt.savefig('bundles_boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```
